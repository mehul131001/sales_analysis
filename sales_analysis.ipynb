### Sales Analysis

#### import necessary libraries


```python
import pandas as pd 
import os
import matplotlib.pyplot as plt
```

#### merging 12 months of sales data into a single csv file

```python
files = [file for file in os.listdir("C:/Users/Mehul/Desktop/sales_analysis/Sales_Data")]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

#### read in updated dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
all_data['Product'].isna().sum()
```

#### clean up the data

```python
all_data[all_data.isna().any(axis=1)]
```

```python
all_data.dropna(axis=0, inplace = True)
```

#### augment data with additional columns

```python
# adding month column
all_data['Month'] = (all_data['Order Date'].str[0:2])

# dropping nan values from 'Month' column
all_data['Month'].dropna(axis=0, inplace = True)

all_data.iloc[517]

#drop rows containing 'Or' value in month
all_data = all_data[~all_data['Month'].str.contains('Or')]
all_data.iloc[516]

#convert all string value in column 'Month' to integer
all_data['Month'] = pd.to_numeric(all_data['Month'])
```

```python
# adding sales column

#converting columns to correct datatype
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data['Sales'] = all_data['Quantity Ordered'] * all_data["Price Each"]
all_data.head()


```

#### best month for sales and the sales amount of that month

```python
results = all_data.groupby('Month').sum()
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

```python
# adding a city column

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

#### city having the highest sales

```python
results = all_data.groupby('City').sum()

#plot
cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation=90, ha='right')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()
```

#### what time should the advertisements be displayed to maximize lilklihood of customer's buying product

```python
# converting the 'Order Date' into data-time format
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()

# fetching the hour from the 'Order Date' column and adding a column named 'hour'
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()

#also adding a column 'Minute'
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()

# plotting hour data
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Orders Placed')
plt.grid()
plt.plot(hours, all_data.groupby(['Hour']).count())

# the plot clearly suggests the advertisements to be around 11am(11) or 7pm(19)
```

#### what products are most often sold together

```python
# spotting rows having same order_id and storing it in a new dataframe
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)

# storing the products with same order_id into a new column named 'Grouped'
df['Grouped']  =df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()

# dropping the duplicated columns with the same value in the group column
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()

#counting the pairs that occurs together the most times
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 4)))

for key, value in count.most_common(10):
    print(key, value)
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
